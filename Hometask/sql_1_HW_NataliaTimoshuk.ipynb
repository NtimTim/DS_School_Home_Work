1. Выберите заказчиков из Германии, Франции и Мадрида, выведите их название, страну и адрес.

```python
pd.read_sql(
    '''
    SELECT * FROM Customers
    where
    Country in ('Germany', 'France') or City = 'Madrid'
    ''',
    con,
)
```

2. Выберите топ 3 страны по количеству заказчиков, выведите их названия и количество записей.

```python
pd.read_sql(
    '''
    SELECT Country, count(*) as cnt
    FROM Customers
    GROUP BY Country
    ORDER BY cnt DESC
    LIMIT 3
    ''',
    con,
)
```

3. Выберите перевозчика, который отправил 10-й по времени заказ, выведите его название, и дату отправления.

```python
pd.read_sql(
    '''
    SELECT S.ShipperName, O.OrderDate FROM
    (SELECT * FROM
     (SELECT * FROM Orders
      order by OrderDate ASC
      LIMIT 10
     )
    order by OrderDate DESC
    LIMIT 1) O
    JOIN Shippers S
    ON O.ShipperId = S.ShipperId
    ''',
    con,
)
```

4. Выберите самый дорогой заказ, выведите список товаров с их ценами.

```python
pd.read_sql(
    '''
    SELECT OMS.OrderId, OMS.summa, P.ProductName, P.Price, OD.Quantity
    FROM
    (SELECT O.OrderId, SUM(OD.Quantity*P.Price) as summa
    FROM Orders O
    JOIN OrderDetails OD ON OD.OrderId = O.OrderId
    JOIN Products P ON P.ProductId = OD.ProductId
    GROUP BY O.OrderId
    ORDER BY summa DESC
    LIMIT 1) OMS
    JOIN OrderDetails OD ON OD.OrderId = OMS.OrderId
    JOIN Products P ON P.ProductId = OD.ProductId
    ''',
    con,
)
```

5. Какой товар больше всего заказывали по количеству единиц товара, выведите его название и количество единиц в каждом из заказов.

```python
pd.read_sql(
    '''
    SELECT P.ProductName, OD.OrderId, OD.Quantity
    FROM
    (SELECT OD.ProductId, SUM(OD.Quantity) as summaQ
     FROM OrderDetails OD
     GROUP BY OD.ProductId
     ORDER BY summaQ DESC
     LIMIT 1) PPS
    JOIN OrderDetails OD ON PPS.ProductId = OD.ProductId
    JOIN Products P ON PPS.ProductId = P.ProductId
    ''',
    con,
)
```

6. Выведите топ 5 поставщиков по количеству заказов, выведите их названия, страну, контактное лицо и телефон.

```python
pd.read_sql(
    '''
    SELECT S.SupplierName, S.Country, S.ContactName, S.Phone
    FROM
    (SELECT P.SupplierId
     FROM Products P
     JOIN OrderDetails OD ON OD.ProductId = P.ProductId
     JOIN Orders O ON O.OrderId = OD.OrderId
     GROUP BY P.SupplierId
     ORDER BY count(O.OrderId) DESC
     LIMIT 5) S5
     JOIN Suppliers S ON S.SupplierId = S5.SupplierId
    ''',
    con,
)
```

7. Какую категорию товаров заказывали больше всего по стоимости в Бразилии, выведите страну, название категории и сумму.

```python
pd.read_sql(
    '''
    SELECT C.Country, Cat.CategoryName ,sum(OD.Quantity*P.Price) as summa
    FROM
    Categories Cat
    JOIN Products P ON P.CategoryId = Cat.CategoryId
    JOIN OrderDetails OD ON OD.ProductId = P.ProductId
    JOIN Orders O ON O.OrderId = OD.OrderId
    JOIN Customers C ON C.CustomerId = O.CustomerId
    WHERE C.Country = 'Brazil'
    GROUP BY C.Country, Cat.CategoryName
    ORDER BY sum(OD.Quantity*P.Price) DESC
    LIMIT 1
    ''',
    con,
)
```

8. Какая разница в стоимости между самым дорогим и самым дешевым заказом из США.

```python
pd.read_sql(
    '''
    SELECT MAX(summa)-MIN(summa)
    FROM
    (SELECT O.OrderId, sum(OD.Quantity*P.Price) summa
     FROM
     Customers C
     JOIN Orders O ON C.CustomerId = O.CustomerId
     JOIN OrderDetails OD ON O.OrderId = OD.OrderId
     JOIN Products P ON OD.ProductId = P.ProductId
     WHERE C.Country = 'USA'
     GROUP BY O.OrderId)
    ''',
    con,
)
```

9. Выведите количество заказов у каждого их трех самых молодых сотрудников, а также имя и фамилию во второй колонке.

```python
pd.read_sql(
    '''
    SELECT count(O.OrderID), E.FirstName||' '||E.LastName as fn, substr(E.BirthDate, 7) || "-" || substr(E.BirthDate,4,2)  || "-" || substr(E.BirthDate, 1,2) as bd
    FROM Employees E
    JOIN Orders O ON O.EmployeeId = E.EmployeeID
    GROUP BY fn
    order by bd DESC
    LIMIT 3
    ''',
    con,
)
```
